```python
import pandas as pd
from scipy import stats

# Load the first CSV file
file_path1 = "data1.csv"
df1 = pd.read_csv('/content/drive/MyDrive/data set/covid-19-confirm-cases2.csv')

# Load the second CSV file
file_path2 = "data2.csv"
df2 = pd.read_csv('/content/drive/MyDrive/data set/covid-19-death-cases1.csv')


# Assuming both CSV files have a column named 'value' representing the data
data1 = df1['Lat']
data2 = df2['Lat']

# Hypothesis test (Two-sample t-test)
t_stat, p_value = stats.ttest_ind(data1, data2)
print(f'Two-sample t-test:\n  t-statistic = {t_stat}\n  p-value = {p_value}')

# ANOVA test (Analysis of Variance)
# You may load a third CSV file for a third dataset if needed
# Assuming both CSV files have a column named 'value' representing the data
f_stat, p_value_anova = stats.f_oneway(data1, data2)
print(f'ANOVA test:\n  F-statistic = {f_stat}\n  p-value = {p_value_anova}')
```

```python
import matplotlib.pyplot as plt
a=t_stat*10
b=p_value*10
c=f_stat*10
plt.scatter(t_stat, p_value, color='black', alpha=0.5)
plt.title("Hypothesis test",size=20)
plt.xlabel("t-statistic",size=20)
plt.ylabel("p-value",size=20)
plt.show()
plt.scatter(f_stat, p_value, color='yellow', alpha=0.5)
plt.title("ANOVA test",size=20)
plt.xlabel("F-statistic",size=20)
plt.ylabel("p-value",size=20)
plt.show()
```
